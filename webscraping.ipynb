```python
import requests
Link="https://www.tatamotors.com/"
Link_text=requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
Soup=BeautifulSoup(Link_text,'lxml')
print(Soup)
```

```python
print(Soup.prettify())
```
