```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[10,20,30,40]
plt.pie(x)
plt.show()
```

```python
y=[10,20,30,60]
mylabels=["Apples", "Bananas", "Cherries", "Dates"]
mycolors=["black", "blue", "yellow", "red"]
myexplode=[0.5,0.2,0,0]
plt.pie(y, labels = mylabels, colors=mycolors, explode=myexplode, shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,6]
colors=["g","r","b","orange"]
plt.bar(x,y,color=colors)
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.title("Bar Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,6]
colors=["g","r","b","orange"]
plt.barh(x,y,color=colors)
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.title("Bar Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,5,6]
colors=["white","white","white","white"]
sizes=[200,300,400,500]
plt.scatter(x,y,color=colors,s=sizes,alpha=1,marker="*",edgecolors="black")
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.title("scatter Plot")
plt.show()
```
